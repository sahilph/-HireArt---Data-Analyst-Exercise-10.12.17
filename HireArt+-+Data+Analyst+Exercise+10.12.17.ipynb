Find which month does the team tend to contact the greatest percentage of its clients?

Analysis Done By: Sahil Phule

```python
import pandas as panda

%matplotlib inline
```

### Load the dataset

```python
df=panda.read_csv("https://docs.google.com/spreadsheets/d/16hLtx8bBDe2GS1MOa3v9hY6Yhm4C30koLoCpiIJ5WDg/export?format=csv")
```

### Check if loaded

```python
df
```

### Check column types

```python
df.dtypes
```

### Convert columm to Date format for ease of handling

```python
df["Date of Contact"] = panda.to_datetime(df["Date of Contact"])
```

### Extract month Get their counts 

```python
df["Date of Contact"].dt.month.value_counts()
```

### Visualize

```python
df["Date of Contact"].dt.month.value_counts().plot(kind="bar")
```

### Conclusion

It seems that the team tend to contact the greatest percentage of clients in the month 10 i.e. **October**
